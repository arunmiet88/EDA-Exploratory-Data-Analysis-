### Data visualisation:

        Data visualisation is the graphical representation of data and information. It use statistical
        graphics,plot,information graph and other tools to communicate information clearly and efficiently.
        Following are commonly used for visualisation
        1- Scatter plot
        2- Box plot
        3- Histogram Plot
        4- Heatmap
        6- Violin plot
        7- Pair Plot
        8- joint Plot

```python
# Scatter Plot
sns.scatterplot(x='Age',y='Fare',hue='Pclass',data=df)
```

```python
# Box Plot
sns.boxplot(x='Pclass',y='Fare',data=df)
```

```python
# Histogram plot
sns.distplot(df['Pclass'],kde=False)
```

```python
# Cat plot
sns.catplot(x="Embarked", y="Survived", hue="Sex",col="Pclass", kind = 'bar',data=df, palette = "rainbow")
```

```python
sns.catplot(x="Embarked", y="Survived", hue="Sex",col="Pclass", kind = 'violin',data=df, palette = "rainbow")
```

```python
# violin plot
sns.violinplot(x='Sex', y='Survived',data=df)
```

```python
# Pair plot
sns.pairplot(df, hue="Sex",diag_kind={'bw': 1} )
```
